**침묵의 봄 (레이첼 카슨 1907~1964)**

**2장,6장,10장,14장**

[**2장.참아야 하는 의무**](#)



  지구 생명의 역사는 생명체와 그 환경의 상호작용의 역사라고 할 수 있다. 넓은 의미로, 지구에 서식하는 동식물의 물리적 형태와 
특성은 환경에 의해 규정된다. 지구 탄생 이후 전체적인 시간을 고려할 때 그 반대 영향, 즉 생물이 주변 환경에 미치는 영향은 
상대적으로 미미하다. 20세기에 들어서 오직 하나의 생물종, 즉 **인간만이 자신이 속한 세계의 본성을 변화시킬 수 있는 놀라운 
위력을 획득했다.** 

인간의 충동적이고 부주의한 활동으로 말미암아 자연의 신중한 속도와는 비교조차 할 수 없는 빠른 속도로 새로운 변화가 초래된다.
인간의 상상력이 고안해내고 실험실에서 만들어진, 그렇기 때문에 **자연 상태에서는 어떤 대응 상대도 없는 합성물질**에도 적응해야 하는 상황이 초래된 것이다.  

생명체가 화학물질에 적응하려면 자연의 척도에 따라 적절한 시간이 필요하다. 이는 그저 인간이 생각하는 몇 년 정도를 의미하는 것이 
아니다. 몇 세대에 이르는 오랜 시간이 필요하다. 설령 기적이 일어나 이런 물질에 쉽게 적응한다고 해도, 실험실에서 계속 새로운 
화학물질들이 꼬리를 물고 쏟아져나올 것이므로 별 성과가 없을 것이다. 

![agriplane](http://www.doopedia.co.kr/_upload/image/1602/26/160226020128670/160226020128670_thumb_1024_w.jpg)
*새로운 화학물질들은 대개 인간이 자연에 대항해 벌이는 전쟁을 위해 만들어진 것이다.
특히, 해충을 없애는 명목하에 제작된 화학제품들은 '해충'만이 아니라 '익충'까지 모든 곤충을 무차별적으로 죽였다.* 

그러나 다윈이 제창한 적자생존론을 증명하듯, **곤충은 살충제에 내성을 지닌 놀라운 종으로 진화해갔다.** 
그러다 보니 이런 곤충에 사용하기 위한 더욱 강력한 살충제가 나오고 그다음엔 이보다 독성이 더 강한 살충제가 등장하는 악순환이 계속되었다.
해충은 살충제 살포 후 생존 능력이 더욱 강해져서 오히려 이전보다 그 수가 많아진다. 따라서 인간은 이 화학전에서 결코 승리를 거두지 못한다.

**해충을 막기 위한 검역과 농약 대량 살포는 그저 시간을 벌기 위한 극도로 비싼 방법에 지나지 않는다.** 

우리가 사활을 걸어야 할 문제는 "그저 한두 종의 식물이나 동물을 억제하는 것"이 아니라고 엘턴 박사는 주장한다. 

그보다는 "자연계의 균형을 유지하고 병충해의 폭발적인 위력과 새로운 공격을 약화시킬 수 있도록" 

**생명체의 특성, 생명체와 환경의 상관관계에 대한 기초 지식을 갖추어야 한다는 것**이다. 

화학물질이 토양, 물, 야생동물과 인간에게 어떤 영향을 미치는지 관련 조사가 이루어지지 않았다는 점도 문제다. 
자연에 닥친 위험을 인식하는 사람은 극소수이다. 전문가의 시대라고 하지만 각자 자신의 분야에서만 위험을 인식할 뿐, 그 문제들이 모두 
적용되는 훨씬 더 광범위한 상황은 인식하지 못하거나 무시한다. 

**공업화 시대라서 그런지 어떤 대가를 치르든 이윤을 올리기만 하면 별다른 제재가 가해지지 않는다.** 

**우리는 이런 잘못된 위안을, 그대로 받아들일 수 없는 사실에 입혀진 당의를 한시라도 빨리 제거해야 한다.** 

해충박멸업자들이 야기한 위험을 책임져야 하는 사람은 바로 일반 시민들이다. 

지금과 같은 방제법을 계속 고집할지 결정을 내리려면 **현재 벌어지는 상황과 진실을 제대로 알아야 한다.** 

장 로스탕(Jean Rostand)은 이렇게 말했다. 

**"참아야 하는 것이 우리의 의무라면, 알아야 하는 것은 우리의 권리다."**

**[의견]**

 저자가 살았던 1960년대는 본격적인 공업화로 인해 한창 무차별적인 개발이 일어났던 시기이다.
말그대로 '환경보존'보다, '산업화, 대량생산'등의 이슈가 더 주목을 받던 때이다. 이는 전 분야에 걸쳐 일어난 변화이며, 물론 농업도 변화의 흐름을 피할 수 없었다. 더 빠르고, 더 많은 생산으로 높은 이윤을 창출하기 위해 화학물질이 무차별적으로 사용되었다. 
저자는 이러한 대세적인 흐름에 문제점을 제기하며 주목을 받았는데, 화학물질이 미치는 영향을 적나라하게 언급하면서 그 위험성을 알리는 한편, 더 큰 문제점으로 지적한 것이 사람들의 '무지'였다.

빠른 성장 탓에 과정에서 발생한 일들이 생략되면서, 사람들 인식속에 결과를 위해선 비정상적인 과정이 당연한것으로 여겨지게 되고, 
그것으로 인해 초래된 결과물로 더 많은 생산물과 이윤뿐만 아니라, 그 이면에 환경파괴가 초래되는 것을 바라보지 못했다. 

저자의 논점은 화학물질의 무조건적인 제한이 아니라, 사용하더라도 **알고 사용하는 것**이 중요하다는 것이다.
이런 점에서 저자가 힘주어 말한 '알아야 하는 것은 우리의 권리다' 라는 말이 더 와닿게 된다.

[**6장.지구의 녹색외투**](#)

물,토양 그리고 지구의 녹색 외투라 할 수 있는 식물들 덕분에 지상에서 동물들이 살아갈 수 있다. 현대인들은 이런 사실을 잘 기억하지 못한다. 
하지만 **태양 에너지를 이용해 우리의 식량을 만들어주는 식물이 없다면, 인간의 존재 자체가 불가능할 것이다.** 그런데 이런 식물에 대해 우리는 정말로 편협한 태도를 보이고 있다. 즉각적인 이용 가치가 있다고 생각하면 그 식물을 잘 키우고 보살핀다. 하지만 지금 당장 별로 바람직하지 않거나 **관심 없는 거라면 즉시 이 식물을 없애버린다.** 

인간에게 해를 끼치는 식물뿐 아니라 먹을거리를 제공해주는 식물이라고 해도 우리의 좁은 소견으로 볼 때 잘못된 시간, 잘못된 장소에 있다면 바로 
제거의 표적이 되는 것이다. 사람들이 별로 원치 않는 식물과 연관이 있다는 이유만으로 제거되는 식물도 있다. 

![제초](http://mblogthumb3.phinf.naver.net/20150616_258/siencia_14344617248032FHje_JPEG/%C1%A6%C3%CA%C0%DB%BE%F7.jpg?type=w2)

화학 제초제는 새 장난감이나 마찬가지고, 이런 제초제들은 신기한 방식으로 작동한다. 
즉 제초제로 사람들은 자연에 대해 경박하게 권위를 행사하고, 제초제 사용에 관해 장기적 인식과 그 불확실한 효과를 언급하면 
'비관론자들의 근거 없는 상상' 으로 무시한다. 수천 군데 마을의 촌로들은 길가의 '잡목'을 싼 값에 없애준다는 살충제 판매원과 열성적인 
방제업자들의 이야기에 솔깃한다. 비용이 잔디 깎는 값보다는 싸다는 말 때문이다. 

그러나 장기적 안목에서 볼때, 화학물질을 대규모로 살포하면 주변 환경은 물론 이 환경에 의지하는 생물들의 건강에도 심각한 손상을 입히므로 
그 금전적인 비용은 훨씬 더 비싸질 것이다. 

가장 널리 사용되는 제초제로는 2, 4-D, 2,4,5-T 그리고 이와 유사한 합성 화학물질 들이 있다. 
이런 물질이 정말 유독한지 아닌지는 논쟁의 여지가 있다. 잔디밭에 2,4-D를 뿌리다가 약물에 젖은 사람들은 심각한 신경염이나 마비로 고생한다. 

1944년 미국에 딱정벌레가 소개된 것은 역사적인 일이었다. 
해로운 식물을 조절하기 위해 그 식물을 먹이로 삼는 곤충을 이용한 북미 최초의 시도였기 때문이다. 
이후 10년간의 노력을 담은 보고서가 1959년 발표되었는데, 고추나물을 조절하기 위한 이런 방식의 접근은 '기대한 것보다 훨씬 더 효율적인 것으로'
밝혀졌다. 고추나물이 예전의 1퍼센트대로 감소했기 떄문이다. 

바람직하지 않은 식물을 방제하는 가장 효과적인 방법은 특정 식물을 먹이로 하는 곤충을 이용하는 것이다. 
하지만 목초지를 관리하는 데 이런 가능성이 대체로 무시되었다. **곤충들은 자신이 원하는 식물만 먹이로 삼는데, 그런 제한적인 식성을 잘 이용
한다면 우리에게 상당한 이익으로 돌아올 것이다**

**[의견]**

저저가 언급한 '세이지'는 저자가 6장에서 이야기하고자 하는 바가 가장 잘 담긴 예라고 생각한다. 
인간의 기준에 의해서 '필요없는, 제거되어야 하는' 식물로 낙인이 찍혀 무분별하게 제거된 세이지는 사실은 생태계에서 동식물들에게 이로운 역할을 
하는 존재가치가 분명한 식물이었다. 

이러한 식물을 인간의 입맛에 따라 제거한 결과는 결국, 연쇄작용으로 나타나 세이지에 의존하던 다른 동식물 마저 사라지고, 생태계가 교란되는 결과를 초래한다. 그리고 이러한 과정에서 사용된 화학제초제는 결국 사람의 건강에도 심각한 영향을 미치는데, 결과론적으로 당장의 결과만을 위한 판단이 장기적으로 도리어 더 좋지 못한 결과로 돌아오게 된다. 이런 점은 비단 6장 뿐만 아니라, 저자가 책의 전반부에 걸쳐 재차 지적하는 부분이다.

바람직하지 못한 식물을 방제하는 방법으로 곤충을 이용하는 방법이 제시되는데, 이는 지속적으로 연구되고 활용되어야 하는 방법이라고 생각한다. 
자연적인 방법으로 생태계를 조정하는 것이 장기적인 안목에서 봤을 때 가장 휴우증이 없는 방법이 될 것이기 때문이다. 
(기회비용에 대한 책정이 적절하게 이루어진다면 경제적인 손실을 충분히 줄이면서 안전한 방법을 택할 수 있을 것이라 생각한다.)  

[**10장.공중에서 무차별적으로**](#)

현재 사용되는 독극물은 예전 그 어떤 것보다 위험한데 놀랍게도 공중에서 무차별적으로 살포되고 있다. 
그리고 구제 목표인 곤충이나 식물뿐만이 아니라 화학약품이 뿌려진 지역에 사는 인간마저도 예기치 못한 재앙처럼 독극물과 접촉하게 되었다. 

숲과 경작지만이 아니라 마을과 도시에도 유독물질이 살포되고 있는 것이다. 

**[의견]**

사람들이 공중 살포에 대해 큰 경각심을 갖게 된 것 중에 하나가 미국 남부에서 진행된 불개미 퇴치 사업인데, 이는 전혀 효과적이지 못했고 
도리어 비용을 더 많이 발생시켰을 뿐만 아니라, 주요 타겟인 불개미를 제거하는데 실패했고, 식물이 아니라 농작물, 가축에게 영향을 미쳤으며 끝내는 사람에게 영향을 미쳤기 때문이다. 공중 살포는 빠른 시간안에 넓은 지역에 살포할 수 있다는 장점이 있으나 그만큼 불특정 다수에게까지 살포될 수 있다는 큰 단점이 있다. 거기에 안전성이 전혀 보장되지 않은 화학물질이 살포되는 것이므로, 그 위험수준은 측정하기 어려웠을 것이다.

너무나도 안타까운 결과이지만(책에서 지속적으로 보여주는 예시들..) 불개미의 경우에도 마찬가지로, 사람들에게 심각한 결과를 초래한다고 알려진 것들이 사실은 생각보다 해를 입히지 않았고, 도리어 이로운 점들이 있다는 것이 밝혀졌다. 저자도 지적했지만, 이는 제초제 사업 업자들과 당시 강한 목소리를 가진 목초업자들의 의해 사실이 왜곡되고, 미디어가 사람들에게 정확한 정보를 제공하지 못한 탓도 있다고 생각한다.  

[**14장.네명 중 한명**](#)

생명이 존재하기 훨씬 전부터 우리를 둘러싼 환경에는 부적절한 요소들이 포함되어 있었다. 
생명체가 등장하고 수백만 년이라는 시간이 흐르면서 셀 수 없이 많은 다양한 생물이 생겨났다. 서두르지 않고 천천히 흐르는 시간, 즉 자연의 시간 동안 생명체는 각종 파괴적인 세력에 적응해갔는데, 제대로 적응하지 못하는 것들은 사라지고 저항력이 강한 것만이 살아남았다. 

이런 자연적 발암물질들은 여전히 악성질환을 유발한다. 하지만 그 수가 적고 생명체 역시 오랫동안 이런 상황에 적응했기에 큰 문제가 없었다. 

**그런데 인간이 등장하면서 상황이 변하기 시작했다. 인간은 생물체 중에서 유독 혼자만 암 유발물질을 인공적으로 만들어낸다.**

인간이 만들어낸 발암물질들은 지난 몇 세기 동안 우리 환경의 일부가 되었다. 그 대표적인 예가 바로 방향족 탄화수소류의 일종인 매연이다.

암의 역사가 오래되었지만 발암물질에 대해 제대로 인식하는 데에도 역시 오랜 시간이 걸렸다. 
지금으로부터 200년 전 한 영국인 의사는 외부 또는 환경적 요인 때문에 악성질환이 발생하는 것은 아닐까 생각했다. 

초기에 발암물질로 추정된 살충물질 중 하나가 바로 비소였다. 
비소는 제초제로 사용되는 비산나트륨과 살충제로 쓰이는 비산칼슘을 비롯한 각종 화학물질에 들어 있다. 

(살충제를 비롯해) 암을 유발하는 화학약품을 피하기란 정말 힘든 일이다. 
두세 종류의 발암물질이 함께 작용하면 그 영향력이 합해져서 더 심각한 문제가 발생한다. 예를 들어 DDT에 노출된 사람이 각종 용매, 페인트 제거제, 탈지제, 드라이클리닝 용매, 마취제 등으로 널리 사용되어 간에 손상을 입히는 탄화수소류 등에도 노출된다고 하자. 그렇다면 DDT의 '안전용량'
이 과연 무슨 의미가 있겠는가? 

암의 해결은 한 번에 한 발짝씩 천천히 진행될 것이다. 암에 대한 치유책을 찾겠다며 몇 백만 달러를 연구에 쏟아붓고 온갖 희망을 투자하면서, 
정작 암을 예방할 수 있는 최고의 기회를 무시하는 일이 일어나서는 안 된다. 

**[의견]**

인간에게 질병으로써 가장 최악의 결과로 치닫는 대표적인 것이 '암'이라고 생각한다. 헌데 이 암의 기원이 불분명한 가운데, 인간만이 생태계에서 유일하게 암 유발물질을 만들어낸다는 것은 가히 충격적이다. 우리가 우리의 목숨을 더 옥죄는 것이기 때문임이 분명하기 때문이다. 

발암물질에 대해 지각하기 시작한 역사가 그리 깊지 않은 것은 분명 현대에 들어서도 여전히 암을 완벽히 해결하지 못한 원인이 된 것이라고 생각한다. 그러나 약 50여년전 저자의 시점(암을 치료하는 의학적 기술이 덜 발달했던 때) 에서 암을 완벽히 치료하는 치료제를 찾는 것보다 암이 애초에 발생하지 않도록 예방하는데 힘쓰는 것이 훨씬 더 효과적이라 주장한다. 나는 이 의견에 전적으로 동의하는데 현실은..이미 반세기가 지났음에도, 미미한 움직임이 있지만, 암의 근원을 차단하기 보다 암이 발생한 후에 효과적으로 제거하는 방법에 더 몰두하고 있기 때문에..안타깝다고 생각한다. 

인간의 관점이 아니라, 자연의 관점에서 바라보고 생태계를 변화시키는 것이 아니라, 생태계를 있는 그대로 보존하려는 노력이 있을 때 비로소 저자가 지적했던 문제점들이 해결된 세상에서 살 수 있지 않을까 생각이 든다. 지난 수세기동안 이는 어쩔 수 없는 선택이었다면, 앞으로 다가올 미래에는 인류의 생존과 직결되는 문제로 위협할 것이기 때문에 저자가 경종을 울렸던 50년전의 일이 앞으로 50년후에는 반복되어 지적되지 않아야 할 것이다.
